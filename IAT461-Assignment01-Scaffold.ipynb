**Checking Timestamps**

```python
#Convert timestamps in sessions data to datetimes
sessions['started_at'] = pd.to_datetime(sessions['started_at'])
sessions['ended_at'] = pd.to_datetime(sessions['ended_at'])

#Convert timestamps in streams data to datetimes
streams['started_at'] = pd.to_datetime(streams['started_at'])
streams['ended_at'] = pd.to_datetime(streams['ended_at'])
```

Looking at our class discussion and at the data files in Google Sheets, some of the timestamps appear to be swapped. This means that the start time is after the end time, and this doesn't make logical sense.

```python
checkReversed = sessions['started_at'] > sessions['ended_at']
print(checkReversed.sum())
```

There are 40 rows with reversed start times and end times. I also want to find out if the computed number of minutes matches the logged minutes.

```python
#Compute number of minutes from the start time and end time
computedMins = (sessions['ended_at'] - sessions['started_at']).dt.total_seconds() / 60

#Check if the computed minutes match the logged duration_mins
matches = ((computedMins[checkReversed].abs() - sessions.loc[checkReversed, 'duration_mins']).abs() < 2).all()
print(matches)
```

It is true that the computed minutes match the logged minutes. Hooray.

```python
sessions.loc[checkReversed, ['session_id', 'started_at', 'ended_at', 'duration_mins']].head()
```

Quick check to make sure this looks okay. Now time to fix the issue.

```python
#For reversed rows, swap the started_at and ended_at values
sessions.loc[checkReversed, ['started_at', 'ended_at']] = (
    sessions.loc[checkReversed, ['ended_at', 'started_at']].values
)

#Check that this worked
remaining = (sessions['ended_at'] < sessions['started_at']).sum()
print(remaining)
```

There are 0 rows left where the end time is before the start time.

**Logged Duration vs Computed Duration**

In streams.csv, there were a few values that raised suspicion, like when I saw the 96 hour maximum duration. That's a long stream. For example, SM00547 has a logged duration of 96 hours, but the start time is 12:00 and the end time is 16:00 on the same day. The logged duration should only be 4 hours.

```python
#Compute number of hours using start and end time
streams['computedHours'] = ((streams['ended_at'] - streams['started_at']).dt.total_seconds() / 3600)

#Compare computed duration with logged duration
wrongDuration = (streams['computedHours'] - streams['stream_duration_hrs']).abs() > 0.1
print(wrongDuration.sum())
```

There are 15 streams in which the logged hours don't match the computed streaming time.

```python
#Rows with wrong durations, showing logged duration vs computed duration
streams.loc[wrongDuration, ['stream_id', 'streamer_id', 'started_at', 'ended_at', 'stream_duration_hrs', 'computedHours']]
```

```python
#Replace logged duration with computed duration

streams.loc[wrongDuration, 'stream_duration_hrs'] = (streams.loc[wrongDuration, 'computedHours'])

#Check that it worked and drop the computedHours column that I made
streams.drop(columns='computedHours')
streams.loc[wrongDuration, ['stream_id', 'streamer_id', 'started_at', 'ended_at', 'stream_duration_hrs']]
```

The stream_duration_hrs that did not match were replaced with the computed time using the start and end timestamps.

For entries that didn't have a reversed start time and end time, I also need to check if computed duration matches logged duration.

```python
#Checking for differences between computed and logged
sessions['computedMins'] = ((sessions['ended_at'] - sessions['started_at']).dt.total_seconds() / 60)
sessions['durationDiff'] = (sessions['computedMins'] - sessions ['duration_mins']).abs()

mismatched = ((sessions['durationDiff'] > 1) & (sessions['computedMins'] >0))

print(mismatched.sum())
```

There are 30 entries in which the computed duration and logged duration differ by over a minute.

```python
#Replacing logged duration with computed duration
sessions.loc[mismatched, 'duration_mins'] = (sessions.loc[mismatched, 'computedMins'].round().astype(int))
```

The mismatched logged minutes have been replaced with computed minutes using the start and end timestamps.

**Heuristics**

*Chat Messages*

When looking at sessions.csv I noticed that the maximum value for the chat_messages column was 1987. I saw this when doing my initial exploration above in this notebook. That's a lot of messages for a person to send in a single session, and when I looked at the entry, the session in which 1987 messages were sent was only 4 minutes long. This is physically impossible. I'm flagging this (and other entries where there are >20 messages per minute logged) as an unusual entry and as potential bot activity.

This would be an artifact, not an error. It's possible that these events really happened, especially if the user was a bot. However, these entries with >20 chat messages/minute are not representative of the behaviour of human viewers and might have to be excluded from analysis if the goal is observing engagement.

*0 Mins with Chats*

As shown below, there appear to be some entries in sessions.csv with 0 logged duration minutes but a number of chats. It's pretty much impossible to send chats if the session lasted 0 minutes.

This is probably an error, especially the entries with a high number of chats but 0 minutes of duration. The timestamp was either recorded at the wrong resolution (like in seconds instead of minutes), or the chat activity is incorrect/in the wrong record. This is unrecoverable information.

```python
#Some of these sessions with 0 minutes recorded still have chat activity
sessions.loc[sessions['duration_mins'] == 0]
```

*Overlapping Sessions*

Looking at timestamps again, there are some overlapping started_at and ended_at times for the same viewer_id on the same day. This shouldn't be possible since I think Twitch only allows a user to watch one stream at a time.

This is likely an artifact, an issue with the measurement/logging method. The end time may have been accidentally logged as the time the stream ended, not when the viewer left. In this case, the duration might be more reliable than the timestamps and will have to be reviewed.
