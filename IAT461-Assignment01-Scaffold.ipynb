### **Dataset Exploration**

**Sessions**

```python
sessions.head()
```

Looking at this data file, I can see that each session, viewer, streamer, and stream has its own ID composed of letters and numbers. There are timestamps for the start and end of the stream. Numeric values are used to express the duration in munutes, number of chats sent, and number of cheers (chats sent with bits, virtual currency). I can see that booleans (True or False) are used to tell whether or not people followed during the stream and if they subscribed during the stream.

```python
sessions.info()
```

There are a total of 22825 entries with a total of 11 columns. There are no null (missing) values here. The values are a mix of data types, including objects (usually strings), ints, and booleans.

```python
sessions.describe()
```

For the numeric data in this file, streams average about 59 minutes with a minimum length of 0 minutes and maximum length of 443 minutes. The average amount of chat messages sent is 6 with a minimum of 0 and maximum of 1987. For cheers, there is an average of 398 with a minimum of 0 and maximum of 10,000. Looking at the entire dataset, it seems like more people are commenting but in smaller amounts of comments. For cheers, there are fewer instances of people sending bits, but when they do, they send a large amount of bits.
